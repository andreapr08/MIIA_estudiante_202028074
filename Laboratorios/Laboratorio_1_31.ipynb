```python
# discrimina variables categoricas , contando la frecuencia de cada una de ellas a partir de la longitud de cada variable 
#add colors
colors = ['#66b3ff','#99ff99','#ffcc99','#ff9999']
def chart(data):
    categoricas  = ["Estado Civil","Escolaridad","Estrato","region","Genero"]
    medias = []
    

    for x in range(len(categoricas)):
        variable = []
        for y in range(len(data[x+1])):
            if data[x+1][y] not in variable:
                variable.append(data[x+1][y])
        lista_medidas = {}
        for k in range(len(variable)):
            medida = data[x+1].count(variable[k])
            lista_medidas[variable[k]] = medida

        medias.append(lista_medidas)

    fig, ax = plt.subplots(3, 2, sharey=False, figsize =(16,16))

    ax[0,0].pie(medias[0].values() ,labels= medias[0].keys(), autopct='%1.1f%%', startangle=90, colors = colors)
    ax[0,1].pie(medias[1].values() ,labels= medias[1].keys(), autopct='%1.1f%%', startangle=90, colors = colors)
    ax[1,0].pie(medias[2].values() ,labels= medias[2].keys(), autopct='%1.1f%%', startangle=90, colors = colors)
    ax[1,1].pie(medias[3].values() ,labels= medias[3].keys(), autopct='%1.1f%%', startangle=90,colors = colors)
    ax[2,0].pie(medias[4].values() ,labels= medias[4].keys(), autopct='%1.1f%%', startangle=90,colors = colors)

    ax[0,0].set_title("Genero")
    ax[0,1].set_title("Estado Civil")
    ax[1,0].set_title("Escolaridad")
    ax[1,1].set_title("Estrato")
    ax[2,0].set_title("region")
     
    return plt.show()

chart(data)
```
